# Distribution plot 
Section 9 - L48  - djp plots 
> sns.distplot()
> sns.jointplot() - between two columns in the dataset 
> sns.pairplot()  - between any 2 possible column combination in the dataset 
> sns.kdeplot()
> sns.rugplot()



```python
import seaborn as sns 
```

```python
 %matplotlib inline # to get the plots inside the cell
```

```python
# Distribution plot - basically a histogram

tips = sns.load_dataset('tips') # Loading tips dataset 
print(tips.head())

sns.distplot(tips['total_bill'], bins=20)  # kde - kernel density estimation , bins - resolution of data

# joint plot - kind (hex, reg, kde)

sns.jointplot(x='total_bill' , y='tip' , data=tips)

# pair plot - hue(categorize the data), palette(color)
sns.pairplot(tips, hue='sex', palette='coolwarm')  # possible combinations of columns 


```

```python
# rug plot 
sns.rugplot(tips['total_bill'])

# kde plot 
sns.kdeplot(tips['total_bill'])
```

# Categorical plot 
Section 9 - L49 - bcb plots - vss plots 
> compares the categorical column with a numerical column
> sns.barplot() - estimator(standard deviation , min etc)
> sns.countplot() - counts the category specified in the dataset 
> sns.boxplot()
> sns.violinplot()
> sns.stripplot()
> sns.swarmplot()
> sns.factorplot()

```python
# categorical plot - bar plot 
import numpy as np 
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.min )

```

```python
# count plot - counts the categorical data in the dataset 
sns.countplot(x='sex', data=tips)
```

```python
# box plot - another categorical plot 
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
# violin plot 
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split =True )
```

```python
# strip plot - jitter- makes more points visible, split - makes the split between male and female 
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', jitter=True, dodge=True)
```

```python
# Swarm plot - combination of strip + violin plot 
sns.violinplot(x='day', y='total_bill', data=tips,dodge=True)
sns.swarmplot(x='day', y='total_bill', data=tips,dodge=True, color='black')
```

```python
# factor plot - generic plot for other categorical plots - replace the kind with the plot needed 
sns.catplot(x='day', y='total_bill', data=tips, kind='box')
```

# Matrix plots - Heatmap 
Section 9 - L50
> convert into a matrix format - use corr / pivot_table
> heatmap - annotates to the values in the matrix form data
> clustermap - clusters similar pattern data in the matrix form data

```python
# datasets used tips , flight 
import seaborn as sns 
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
%matplotlib inline


print(tips.head())
print('-----------------------------------------')
print(flights.head())
print('-----------------------------------------')

# converting the tips data into some sort of matrix using corr 
tc = tips.corr()
print(tc)
print('-----------------------------------------')
sns.heatmap(tc, annot=True, cmap='coolwarm', linewidths=0.5, linecolor='black')




```

```python
# converting the flight data into some sort of matrix using pivot_table
fp = flights.pivot_table(index='month', columns='year', values='passengers')
print(fp)
print('-----------------------------------------')
sns.heatmap(fp, cmap='magma', linewidths=1.5, linecolor='black')


# cluster map 
sns.clustermap(fp, cmap='coolwarm', standard_scale=1 )
```

# Grids 
Section 9 - L51
> Create the grid and then plot different types to the grids 
> pairgrid 
> facetgrid 


```python
import seaborn as sns 
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')
print(iris.head())
iris['species'].unique()

# use pairplot 
sns.pairplot(iris)

# use PairGrid - to get empty grids (subplots basically)

pg =sns.PairGrid(iris)
# map any plot to those grids 
# pg.map(sns.barplot)  # mapping the entire grid system 
pg.map_diag(sns.distplot) # map only the diagonal 
pg.map_upper(sns.scatterplot) # map only the upper diag 
pg.map_lower(sns.kdeplot)   # map only the lower diag

# use FacetGrid
fg = sns.FacetGrid(tips, row='smoker', col='time') # creates the grid/ subplots

# map the grid to a specific plot 

# fg.map(sns.distplot,'total_bill')

# map the grid to another plot 
fg.map(sns.scatterplot, 'total_bill', 'tip')



```

# Regression plots 
Section 9 - L52
> lmplot 

```python
# lmplot 

# approach 1 - using hue 
print(tips.head())
sns.lmplot('total_bill','tip', tips, hue='sex', markers=['o', 'v'], scatter_kws={'s':50} )

# approach 2 - col , row
sns.lmplot('total_bill', 'tip', tips, row='time', col='day', aspect=0.6, hue='sex')
```
