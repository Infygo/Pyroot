```python
# Perceptron model 

# Single neuron - Perceptron - Multilayer perceptron model - DL neural network 
# Math Concepts - Activation functions, backpropagation and Gradient descent 
# neuron working - inputs from multiple dendrites > processing in nucleus > outputs in axon 
# Perceptron model - inputs x1, x2 > perceptron (neuron) > output 
# improvise perceptron model - inputs x1, x2 > adjust weights w1, w2 to x1 x2 > perceptron(neuron)> output 
# What if x1 / x2 is 0 > then weights doesnt matter anyways and how to adjust/improvise the model?
# Bias > x1*w+b x2*w+b > perceptron > output (y)
```

```python
# Neural networks (Multilayer perceptron model)

# Fully connected - output of first layer serves as the input of the 2nd layer and so on 
# 1st layer - input layer (tabular data - data to be predicted)
# last layer - output layer (can have more than one neuron)
# intermediate layer - hidden layers (highly interconnected and distance away from know input/output values)
# Deep neural network - 2 or more hidden layers 
# width of network - how many neurons are there in a layer 
# Depth of network - how many layers are there in total 
# Classification tasks needs output between 0 and 1 
# We'l use activation functions to set boundaries to output values from the neuron 
```

```python
# Activation functions - limiting the value of output 

# x*w + b => w represents the weight/ importance of that input 
# b represents the offset value , x*w has to reach a threshold value before b making an impact 
# b = 10 , x*w wont have an impact until becomes 10 and after that it depends on w 
# z = wx +b 
# f(Z) - tensor input consisting of multiple values 
# activation function 1 -Sigmoid function -  output bound [0,1]
# activation function 2 -Hyperbolic tangent - output bound [-1,1]
# activation function 3 - Rectified Linear unit - max(0,z) => o/p less than zero =>0, greater than 0 actual value 
# ReLu - Rectified linear Unit - overall good performance because of handling vanishing gradient 

# https://en.wikipedia.org/wiki/Activation_function
```

```python
# Multi class classification considerations 

# Output layer has multiple neurons 
# Mutli class situations 
# - Non exclusive classes - data point can have multiple classes assigned to it 
# - Mutually exclusive classes - one data point has only one class 

# Non exclusive classes - example tag in a group photo 
# Mutually exclusive class - example of either being a bw photo or color photo

# Mutually exclusive class 
# - Datapoint1 - red, Datapoint2 - blue, Datapoint3 - green etc 
# Build a classification matrix [one high encoding] - Dataframe structure Datapoints rows , colors as colums 
# 1 for datapoint 1 under red column and others 0's
# Activation function - softmax function 
# softmax function - calculate the probabilities of each target class over all possible target classes 
# sum of the probabilities = 1 
# [Red, blue, green] => [0.1,0.6,0.3] => target class will be blue 

# Non exclusive class 
# same as above except 1 can hold true for multiple columns 
# Activation function - sigmoid function / logisitic function - because it makes a single data point fed into 
# the function to have multiple classes assgined to it 

# Say for IRIS data set you have three classes and your inputs can belong to only one type of flower class. 
# So, how can we ever have a case where mutually exclusive classes can have same probabilities?
# its how sure the model is of belonging to a class
```

```python
# How networks learn the weight and bias automatically => Cost functions and gradient descent 

# Cost functions / loss functions / error functions 
# Cost function - must be an average so it can output a single value - measures how far off the model is 
# from the true value 
# variables 
# y - represent the true value 
# a - neuron prediction value 
# z = w*x + b 
# Apply z into an activation func σ(z) = a [***]

# Quadratic cost func - most common cost function 
# C = 1/2n Σ ||y(x) - al(x)||^2
# y - true value, a(x) - predicted value / activation function output of l(last) layer [***]
# square in the cost function - makes everything positive and punishes large error 10$
# C(w, b , input of single training sample, output of single training sample) [***]


# How to find the weight that will have low cost 
# weights n dimensional and 1000's of weight => solve it with gradient descent method 

# Gradient descent method 
# > Remember the simple graph model with a U parabola having its min @ wmin which minimises the cost function 
# > Calculate the slope and move in the downward direction of slope 
# > keep repeating the process until we converge to zero indicating minimum
# > change step size => smaller step size => longer time to find the min 
# > larger step size => we risk overshooting the minimum
# process of figuring step size to find the min weight values  - learning rate 
# step size was equal irrespective of being small or large 
# adaptive stepsize = changing stepsize dynamically => adaptive gradient descent 

# Adaptive gradient descent (ADAM ) - super optimised and fast 
# > start with larger stepsize and go smaller as the slope gets closer to zero 

# N dimensional vectors (tensors) => we use gradient instead of a derivative 
# ∇C(w1,w2,....wn)

# For classification problems we'l use cross entropy loss function instead of Quadratic loss function 
```

```python
# Backpropogation - propogate backwards in a neural network and update the weights and biases 

# Imagine with one neuron per layer mode 
# O -w1+b1 - O -w2+b2 - O -w3+b3 - O -w4+b4

# O(L-n) O(L-2) O(L-1) O(L) 

# Focus on the last 2 layers  L , L-1 
# x is technically an input only at the very first layer 
# as we go inside the network it would become the output of a prev layer to the next layer => w*a + b 
# which means 
# Z^L = w^L a^L-1 + b^L
# a^L = σ(z^L)
# C = (a^L -y)^2

# Understand how sensitive is the cost function C is to changes in w
# chain rule - ∂Co/∂wl = ∂zl/∂wl * ∂al/∂zl * ∂Co/∂al

# understand how cost function changes w.r.t to bias 
# ∂Co/∂bl = ∂zl/∂bl * ∂al/∂zl * ∂Co/∂al

# use the gradient to go back through the network and adjust the weight and biases to minimise the output of 
# error vector on the last output layer 

# Hadamard product - element by element multiplication - 


# Process for training neural network
# Step1 : Use input x and set the activation function a for the input layer 
# => z = wx+b , a = σ(z)
# resulting a then feeds into the next layer and so on 
# Step2:
# compute z and a's for each layer zl = wlal-1 + bl, al = σ(zl)
# Step3:
# Compute error vector = δL = ∇(subscript)aC ⊙ σ(zL)
# ∇(subscript)aC = (a^L -y)
# Step4:
# δl = (wl+1)^T δl+1 ⊙ σ'(zl)

```

```python
# Keras syntax basics 
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 

df = pd.read_csv('../TensorFlow_FILES/DATA/fake_reg.csv')
df.head()
```

```python
# Explore the dataset 
sns.pairplot(data=df)
```

```python
# Predict the price based on feature1 and feature2 => X = features , y = price 

from sklearn.model_selection import train_test_split
X = df[['feature1', 'feature2']].values       

# tensor flow accepts in numpy array rather than dataframe

y = df['price'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape   # 70 %
```

```python
X_test.shape    # 30%
```

```python
# Normalize and scale feature data , only the features 
from sklearn.preprocessing import MinMaxScaler
# help(MinMaxScaler)
mmscaler = MinMaxScaler()
```

```python
# fit the train data 
mmscaler.fit(X_train)
```

```python
# transform the X_train
X_train = mmscaler.transform(X_train)
```

```python
# transform the X_test
X_test = mmscaler.transform(X_test)
```

# Part 2 

```python
# working from a virtual env tf_gpu 
from tensorflow.keras.models import Sequential # neural model 
from tensorflow.keras.layers import Dense # layer type - where every neuron is connected to the other neuron 
```

```python
help(Sequential)
```

```python
# Approach 1 building a model
model = Sequential(layers=[Dense(4,activation='relu'),  #1st layer 4 neurons, relu activation fn 
                           Dense(2,activation='relu'),  # 2nd layer 2 neurons, relu act fn 
                           Dense(1)]) # output layer 

```

```python
# Approach 2  - will be using this 
model = Sequential()
model.add(Dense(4,activation='relu'))  # 1st layer 
model.add(Dense(4, activation='relu')) # 2nd layer 
model.add(Dense(4, activation='relu'))

model.add(Dense(1))                    # output layer - predicts the price and we'l compare it with the true price 


```

```python
# checkout keras notebook for the loss func
# multi class classification prob - categorical_crossentropy 
# binary classification prob - binary_crossentropy 
# Regression problem - mse

model.compile(optimizer='rmsprop',loss='mse')
```

```python
# fit the model 
model.fit(x=X_train, y=y_train,epochs=250)   # epoch - going through the dataset (X_train)
```

```python
# plot the mse 
loss_df = pd.DataFrame(model.history.history)
loss_df.head()
```

```python
loss_df.plot()
```

# Part 3 - Evaluate the model 

```python
model.evaluate(X_test,y_test,verbose=0)

# mse of 26.70 
```

```python
model.evaluate(X_train, y_train,verbose=0)
# mse of 24.50
```
