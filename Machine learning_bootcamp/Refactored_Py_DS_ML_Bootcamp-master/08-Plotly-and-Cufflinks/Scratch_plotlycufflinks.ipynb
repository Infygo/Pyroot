# Plotly & Cufflinks
> Plotly - interactive vis library 
> cufflinks - connects the plotly with the pandas 

```python
# imports needed for the plotly and cufflinks libs 

import pandas as pd 
import numpy as np 

from plotly import __version__ 
import cufflinks as cf 
# print(__version__) #4.4.1
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
init_notebook_mode(connected=True)   # connects the js to the notebook mode 

cf.go_offline() # having cufflinks in offline mode() 
# cf.set_config_file(offline=False, world_readable=True)

```

```python
# creating datasets 
df1 = pd.DataFrame(np.random.randn(100,4),columns=['A','B','C','D'])
# df1.head()
df2 = pd.DataFrame({'Category':['A','B','C'],'Values':[32,43,50]})

# plot using plotly and matplot 

df1.plot()
df1.iplot()
```

```python
# using plotly for different plottypes 

df1.iplot(kind='scatter',x='A',y='B',mode='markers',size=5 )
```

```python
# use a bar plot - 
print(df2)
df2.iplot(kind='bar',x='Category', y='Values', colors='blue')
```

```python
# iplot with aggregate funcs - count 
df1.count().iplot(kind='bar')  # count of the values under A, B, C, D Columns 

```

```python
df1.sum().iplot(kind='bar')
```

```python
# box plot 
df1.iplot(kind='box')
```

```python
# 3d Plots surface plot 
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[5,4,3,2,1]})
df3.iplot(kind='surface',colorscale='rdylbu')
```

```python
# histogram 
df1.iplot(kind='hist',bins = 25)
```

```python
# spread plots 
df1[['A','B']].iplot(kind='spread')
```

```python
# Bubble plot 
df1.iplot(kind='bubble',x='A',y='B', size='C')
```

```python
# scatter plot - similar to pair plot in seaborn 
df1.scatter_matrix()
```
