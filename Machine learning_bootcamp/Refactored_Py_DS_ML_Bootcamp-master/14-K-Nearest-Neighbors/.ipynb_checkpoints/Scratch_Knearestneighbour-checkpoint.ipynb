# KNN - K nearest neighbour 
> Chapter 4 - Intro to stat learning 
> KNN
> Consider a horse or dog sample plot - since all the neighbouring points will be considered and influences the result 
> KNN
  > Training algorithm - stores all the data 
  > Prediction algorithm - calc the distance from x(new data point) to all other data points 
                         - sort the points in data with increasing distance from x 
                         - predict the majority label of K closest points 
  
> K will impact which class the new data point belongs to 
> Pros - Simple , easy to add more data, less parameters (K, Distance metric) 
> Cons - High prediction cost, categorical features dont work well, not good with high dimensional data 

> Mostly used in interview datasets > no complete understanding of the data

# KNN with python

```python
# imports library 
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 

# read the classified data 
df = pd.read_csv('Classified Data')
df.head()
```

```python
# using standard scalar 
from sklearn.preprocessing import StandardScaler     # standard scaler from preprocessing 
scaler = StandardScaler()   # instantiate the scaler 


```

```python
# fit the scaler 
scaler.fit(df.drop(labels=['TARGET CLASS','Unnamed: 0'], axis=1))    # dropping the target class column alone and fit it 
```

```python
# transform the scaler to get the features 
scaled_features = scaler.transform(df.drop(labels=['TARGET CLASS','Unnamed: 0'], axis=1))   # features using transform
```

```python
df.columns
```

```python
# convert the scaled_features into a dataframe 
df_feat = pd.DataFrame(data=scaled_features, columns=['WTT', 'PTI', 'EQW', 'SBI', 'LQE', 'QWG', 'FDJ', 'PJF',
       'HQE', 'NXJ']) # assigning the featured data with column names except target class col name 

```

```python
df_feat.head()   # this standarized dataset will now represent the distance between each feature 
```

```python
# train test split - X represents the labels used to predict, y represents the label that will be predicted 
from sklearn.model_selection import train_test_split
X = df_feat
y = df['TARGET CLASS']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)   # instantiate the model
```

```python
knn.fit(X_train, y_train)   # fit and train the model 
```

```python
pred = knn.predict(X_test)   # predict 
```

```python
# Evaluate the model 
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(classification_report(y_test,pred))
print(confusion_matrix(y_test, pred))
```

```python
# choosing a correct K value 

Error_rate = []   # list to store the error rate 

for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    print(pred_i)
    Error_rate.append(np.mean(pred_i!=y_test))

```

```python
Error_rate
```

```python
# Lets plot the Error rate to have a better understanding 
plt.figure(figsize=(10,6))
plt.plot(range(1,40), Error_rate, color='blue', ls='--', marker='o',markersize=10,markerfacecolor='r')
plt.xlabel('K value')
plt.ylabel('Error rate')
plt.title('K value vs Error rate')
plt.grid(True)
```

```python
# lets do the metric for k with higher values 
knn = KNeighborsClassifier(n_neighbors=17)
knn.fit(X_train, y_train)
pred = knn.predict(X_test)

# metrics 1
print('Metrics with K=17')
print(classification_report(y_test, pred))
print(confusion_matrix(y_test,pred))


knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
pred = knn.predict(X_test)


# metrics 2
print('Metrics with K=1')
print(classification_report(y_test, pred))
print(confusion_matrix(y_test,pred))

```
