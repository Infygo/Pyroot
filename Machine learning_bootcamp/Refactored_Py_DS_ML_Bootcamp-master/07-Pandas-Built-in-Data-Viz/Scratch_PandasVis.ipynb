```python
import numpy as np 
import seaborn as sns 
import pandas as pd 
%matplotlib inline 
import matplotlib.pyplot as plt

#plt.figure("whitegrid")

df1 = pd.read_csv("df1", index_col=0)
print(df1.head())

df2 = pd.read_csv("df2")
print(df2.head())

# approach 1
df1['A'].hist(bins=25)
# approach 2 
df1['A'].plot(kind='hist', bins=50)
# approach 3 
df1['A'].plot.hist(bins=25)


# area plot 
df2.plot.area(alpha=0.25)

# bar plot 
df2.plot.bar(stacked=True)

# line plot 
# df1.plot.line(df1.index, 'B')
```

```python
# inbuilt panda data viz plots - scatter plot 
df1.plot.scatter(x='A',y='B',c='C',colormap='viridis')
```

```python
# using an inbuilt panda box plot 
df2.plot.box()
```

```python
# hexbin plot 
df_random = pd.DataFrame(np.random.rand(1000,2),columns=['a','b'])
print(df_random.head())
df_random.plot.hexbin(x='a',y='b',gridsize=25,colormap='viridis')
```

```python
# kde / density plot 
print(df2.head())
df2.plot.density()
```
