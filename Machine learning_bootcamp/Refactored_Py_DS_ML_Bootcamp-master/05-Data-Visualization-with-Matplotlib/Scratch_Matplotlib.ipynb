```python
# matplotlib.org - resource site 
import matplotlib.pyplot as plt 
# %matplotlib inline    # no need to use plt.show()
%config IPCompleter.greedy=True 

import numpy as np 
x = np.linspace(0,5,11)
y = x ** 2


# Functional way 
plt.plot(x,y, 'r--')
# plt.show()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Title')

```

```python
# subplot 

plt.subplot(1,2,1)  # no of nrows, ncols, 1st subplot
plt.plot(x,y,'r')   # plot the subplot 1

plt.subplot(1,2,2) # 2nd subplot 
plt.plot(y,x,'b')  # plot the subplot 2 
```

```python
# OOPS - matplotlib 
# create a figure (canvas) - add an axis to the figure (l, b, w, h) - plot it on the axis 

# create a figure 
fig = plt.figure()

# add an axis to the figure 
axis = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 
axis.set_xlabel('X')
axis.set_ylabel('Y')
axis.set_title('Title')

axis.plot(x,y)
```

```python
#  oops 
# create a canvas 
fig = plt.figure() # creating a canvas / figure 

# main axis 
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])  # left , bottom , width, height - needs to be 0 and 1 
# inset axis 
axis2 = fig.add_axes([0.2,0.5,0.4,0.3])

axis1.plot(x,y)
axis1.set_title('Large plot')

axis2.plot(y,x)
axis2.set_title('Small plot')

```

# Matplotlib Part 2 
> plt.tightlayout - avoids overlap 
> fig, axes = plt.subplot(nrows, ncols, subplot#)
> iterate through subplot, access it 

```python
# Subplots using OOPS 
fig,axes = plt.subplots(nrows=1, ncols=2) # basically the same - setting the canvas/figure in one shot 


# iterating through the subplots 

# for current_ax in axes:
#     current_ax.plot(x,y)

# accessing the subplot 

axes[0].plot(x,y,'r-')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('1st plot')
axes[1].set_title('2nd plot')
axes[1].plot(y,x)

# creating multiple subplots 
fig,axes = plt.subplots(3,3)
plt.tight_layout()
```

# Figure size and DPI - Save plot 
> use it on subplot , normal oops plot 
> savefig method - to save the plot/figure 
> legends - add label to the axes - add legend to the axes 
> figsize and DPI - X axis, y axis length 
> plot apperance - color, linewidth, linesize, alpha(transparency), marker, markersize, markerfacecolor, markeredgewidth, markeredgecolor, 
> plot control - xlim, ylim



```python
# Figsize and DPI 

fig = plt.figure()  # create a canvas 
axes = fig.add_axes([0,0,1,1]) # adding an axis to canvas - l, b, w, h
axes.plot(x,y)
axes.set_title('Normal plot')


# using figsize and DPI
fig1 = plt.figure(figsize=(3,3))  # figsize, dpi 
axes1 = fig1.add_axes([0,0,1,1])
axes1.plot(x,y)
axes1.set_title('Figsized plot')

# use figsize and DPI with subplots 

fig2,axes = plt.subplots(nrows=2,ncols=1, figsize=(4,2)) # 2 rows and 1 column 
axes[0].plot(x,y)
axes[0].set_title('subplot1')
axes[1].plot(y,x)
axes[1].set_title('subplot2')
plt.tight_layout() # avoid overlapping 

fig2.savefig('myplot.png', dpi=200)

# Legends 
fig3 = plt.figure() #canvas 
axes = fig3.add_axes([0,0,1,1])
axes.plot(x,x**2, label='squared')
axes.plot(x,x**3, label='cubed')

axes.legend(loc=10)  # loc specifies the position of the legend 

# Plot apperance 
# color, linewidth, alpha(transparency), linestyle, markers , markersize, markerfacecolor, markeredgewidth, 
# markeredgecolor

# create a canvas 
fig4 = plt.figure()

# add an axes 
axes = fig4.add_axes([0,0,1,1]) # adding an axis 

axes.plot(x,y, color='purple', linewidth=2.5,alpha=0.75, linestyle='-.', marker='o', 
          markersize=10, markerfacecolor='yellow', markeredgewidth='2.5', markeredgecolor='blue')

# xlim , ylim 
axes.set_xlim([0,1])  # limits the plot to x axis 0-1 
axes.set_ylim([0,1])  # limits the plot to y axis 0-1 



```

```python
# setting xscale / yscale 
# set_xscale , set_yscale 

import matplotlib.pyplot as plt
import numpy as np 
%config IPCompleter.greedy=True 

# create a subplot 
fig, axis = plt.subplots(1,2,figsize=(10,5))
x = np.linspace(0,5,11)

# subplot 1 configuration 
axis[0].plot(x,x**2, x, np.exp(x))
axis[0].set_title('Subplot1')
axis[0].set_xlabel('X axis')
axis[0].set_ylabel('Y axis')

# subplot 2 configuration 
axis[1].plot(x, x**2, x, np.exp(x))
axis[1].set_yscale("log")
axis[1].set_xlabel('X axis')
axis[1].set_ylabel('Y axis')
axis[1].set_title("Subplot2")



```

```python
# setting ticks and tick labels 

import matplotlib.pyplot as plt
import numpy as np 

x = np.linspace(0,5,11)

# subplots - adding ticks 
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x, x**2,label='Squared', lw=1.5, marker='x', ls='dashed')
ax.plot(x,x**3,label='Cubed', lw=1.5, marker='o', ls=':')
ax.legend(loc=10)
ax.set_xticks([1,2,3,4,5])

# set_xtickslabels - doesnt seem to be working ??/

#ax.set_xtickslabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'])
```

```python


# Scientific notations 

# Import ticker - formatter - set_scientific, set_powerlimits, set_major_formatter

import matplotlib.pyplot as plt 
import numpy as np 
x = np.linspace(0,5, 11)
fig, ax = plt.subplots(1,1,figsize=(10,5))
ax.plot(x,x**2)

# Lets use scientific notation to the axis 
from matplotlib import ticker
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True)
formatter.set_powerlimits((-1,1))
ax.yaxis.set_major_formatter(formatter)
```

```python
# axis numbers and axis label spacing - padding 
# pad the matplot x ticks first - pad the x and y axis next 

import numpy as np 
import matplotlib.pyplot as plt 

fig, ax = plt.subplots(1,1)

ax.plot(x,x**2, x, np.exp(x))

# distancing using pad between labels and axis numbers 

ax.xaxis.labelpad = 50
ax.yaxis.labelpad = 50

plt.rcParams['xtick.major.pad'] = 10
plt.rcParams['ytick.major.pad'] = 10

# distance between the labels and the axis numbers 
# axis labels 
ax.set_xlabel('X')
ax.set_ylabel('Y')
# axis numbers 
ax.set_xticks([0.5,1.5,2.5,3.5,4.5]) # set ticks at the places in the list 

# restore to defaults 
plt.rcParams['xtick.major.pad']= 3
plt.rcParams['ytick.major.pad']= 3




```

```python
# adjust feature while saving 
# fig.subplots_adjust - keyword 
import numpy as np 
import matplotlib.pyplot as plt 

x = np.linspace(0,5,11)

# using functional plot and axis 
# fig = plt.figure()
# axis = fig.add_axes([0.1, 0.1, 0.8,0.8])
# axis.plot(x,x**2, x, x**3)

# using the subplots 
fig, ax = plt.subplots(1,1)
 

ax.set_yticks ([0,50,100,150]) # set the ticks along y axis 
ax.set_title('Adjust the plot')
ax.set_xlabel('x')
ax.set_ylabel('y')

ax.plot(x, x**2, x, np.exp(x))  # plot

fig.subplots_adjust(left=0.50, right=0.9, top=0.9, bottom=0.1)



```

```python
# Grid appearance 
import numpy as np 
import matplotlib.pyplot as plt 

x = np.linspace(0,5,11)
# create a subplot with 2 plots 
fig, ax = plt.subplots(1,2,figsize=(7,3)) 
ax[0].plot(x,x**2, x,x**3)

ax[0].grid(True)
ax[0].set_title('Def gridding')

ax[1].plot(x, x**4, x, x**5)
ax[1].grid(color='g', alpha=0.75, ls='dashed', lw=0.75)
ax[1].set_title('Gridding')


```

```python
# Axis spines 
import numpy as np 
import matplotlib.pyplot as plt 

x = np.linspace(0,5, 11)
fig,ax = plt.subplots(1,1,figsize=(5,3))

ax.spines['left'].set_color('blue')
ax.spines['right'].set_color('blue')
ax.spines['left'].set_linewidth(3)

ax.spines['top'].set_color('red')
ax.spines['bottom'].set_color('red')
ax.spines['bottom'].set_linewidth(3)

ax.set_title('Axis spines')


```

```python
# twinx , twiny 
import numpy as np 
import matplotlib.pyplot as plt 

x = np.linspace(0,5, 11)


fig, ax1 = plt.subplots()
ax1.plot(x,x**2,lw=2, color='b')
ax1.set_ylabel(r"Area$(m^2)$",fontsize=18, color='b') # setting y label to blue color 
# setting color to the axis ticks 
for label in ax1.get_yticklabels():
    label.set_color('blue')

# use the twinx - way of making sure both the plots have common x axis 
ax2 = ax1.twinx()
ax2.plot(x, x**3,lw=1.5,color='g')
ax2.set_ylabel(r"Volume$(m^3)$", color='g',fontsize=18)
for label in ax2.get_yticklabels():
    label.set_color('g')


```

```python
# complex cubic plots 
# set the xaxis ticks first - bottom > set the spines at bottom at 0 position 

fig, ax = plt.subplots()
ax.xaxis.set_ticks_position("bottom")    # setting x axis ticks at the bottom
ax.spines['bottom'].set_position(('data', 0)) # setting x axis spines at bottom to 0 

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0)) 

xx = np.linspace(-0.75, 1, 100)  # linearly splits 100 spaces between .75, 1
print(xx)
ax.plot(xx, xx**3)  
```
